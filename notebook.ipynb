### Cleaning Data

```python
# Menghapus Data yg kosong
df.dropna(inplace=True)
# Hasil setelah Cleaning data Kosong
df.isnull().sum()
```

Mengahapus Outliers Utk Data Numerik

```python
numeric_data = df.select_dtypes(include=[np.number])

for coluumn in numeric_data.columns:
  Q1 = numeric_data[coluumn].quantile(0.25)
  Q3 = numeric_data[coluumn].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  numeric_data = numeric_data[(numeric_data[coluumn] >= lower_bound) & (numeric_data[coluumn] <= upper_bound)]
  # Menampilkan Hasil nya
numeric_data.describe()
```

**Insight:**
- Cleaning Data yg kosong
- Menhilangkan Outliers nya

## Exploratory Data Analysis (EDA)

Utk Tahap Exploratory Data Analysis (EDA) ini saya ingin Menanyakan Dari Pertanyaan Bisni Yaitu:

- Kami mengeksplorasi nilai rata-rata untuk panjang nama produk, panjang deskripsi produk, berat, dan dimensi total per kategori produk. Wawasan diperoleh untuk mengidentifikasi kategori mana yang biasanya memiliki produk yang lebih besar dan lebih berat.

-  Kami menganalisis distribusi berat produk dan dimensi berdasarkan kategori. Visualisasi seperti boxplot digunakan untuk memahami apakah produk yang lebih berat dan lebih besar cenderung berada pada kategori tertentu. Ini dapat memberikan wawasan tentang bagaimana dimensi produk memengaruhi harga.



## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Menambahkan fitur total dimensi
df['total_dimension'] = df['product_length_cm'] + df['product_height_cm'] + df['product_width_cm']

# Menghitung rata-rata per kategori
numeric_columns = df.select_dtypes(include=['float64']).columns  # Memilih kolom numerik
category_stats = df.groupby('product_category_name')[numeric_columns].mean()

# Pertanyaan 1: Kategori produk mana yang memiliki produk lebih besar dan lebih berat?

category_stats_sorted_by_weight = category_stats.sort_values(by='product_weight_g', ascending=False)
category_stats_sorted_by_dimension = category_stats.sort_values(by='total_dimension', ascending=False)



# Plot untuk visualisasi
fig, ax = plt.subplots(figsize=(10, 6))
category_stats[['product_weight_g', 'total_dimension']].plot(kind='bar', ax=ax)
plt.title('Rata-rata Berat dan Dimensi Total per Kategori')
plt.ylabel('Nilai Rata-rata')
```

### Pertanyaan 2:

```python
# Plot hubungan antara kategori dan berat produk
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='product_category_name', y='product_weight_g')
plt.xticks(rotation=90)
plt.title('Distribusi Berat Produk Berdasarkan Kategori')
plt.show()

# Plot hubungan antara kategori dan dimensi produk (Panjang, Tinggi, Lebar)
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(data=df, x='product_category_name', y='product_length_cm', ax=ax[0])
ax[0].set_title('Panjang Produk Berdasarkan Kategori')
ax[0].tick_params(axis='x', rotation=90)

sns.boxplot(data=df, x='product_category_name', y='product_height_cm', ax=ax[1])
ax[1].set_title('Tinggi Produk Berdasarkan Kategori')
ax[1].tick_params(axis='x', rotation=90)

sns.boxplot(data=df, x='product_category_name', y='product_width_cm', ax=ax[2])
ax[2].set_title('Lebar Produk Berdasarkan Kategori')
ax[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

## Conclusion

1. Berdasarkan analisis, kategori produk yang memiliki dimensi total lebih besar dan berat lebih tinggi adalah kategori dengan produk yang lebih besar dan lebih berat. Kategori produk ini berpotensi memiliki biaya produksi dan pengiriman yang lebih besar.
2. Dapat diasumsikan bahwa produk dengan dimensi dan berat yang lebih besar mungkin cenderung lebih mahal, meskipun ini perlu verifikasi lebih lanjut melalui analisis harga.
